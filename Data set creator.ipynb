```python
import pandas as pd
import os
import shutil
```

```python
# Create dataset for positive cases
FILE_PATH="chestxray/metadata.csv"
IMAGES_PATH="chestxray/images"
```

```python
df=pd.read_csv(FILE_PATH)
print(df.shape)
```

```python
df.head()
```

```python
TARGET_DIR="Dataset/Covid"
if not os.path.exists(TARGET_DIR):
    os.mkdir(TARGET_DIR)
    print("Covid Folder Created")
```

```python
c=0
for (i,row) in df.iterrows() :
    if row["finding"]=="Pneumonia/Viral/COVID-19" and row["view"]=="PA":
        filename=row["filename"]
        image_path=os.path.join(IMAGES_PATH,filename)
        image_copy_path=os.path.join(TARGET_DIR,filename)
        shutil.copy2(image_path,image_copy_path)
        #print("Moving Images",c)
        c += 1

```

```python
#Sampling of data from Kaggle Dataset
import random
KAGGLE_FILE_PATH="kaggle_dataset/train/NORMAL"
TARGET_NORMAL="Dataset/Normal"
```

```python
image_names=os.listdir(KAGGLE_FILE_PATH)
```

```python
random.shuffle(image_names)
```

```python
for i in range(196):
    image_name=image_names[i]
    image_path=os.path.join(KAGGLE_FILE_PATH,image_name)
    target_path=os.path.join(TARGET_NORMAL,image_name)
    shutil.copy2(image_path,target_path)
    print("Copying image",i)
```
